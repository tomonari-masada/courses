## 3-5 datetimeモジュール

```
import datetime
```

特定のデータ型について、

その実体（実例）を一つ作るメソッドは、

そのデータ型と同じ名前になっている。

date型は、日付のみを扱うことができる。

```
day = datetime.date(1970, 6, 26)
```

```
print(day)
```

```
day.weekday()
```

### 3-5-3 datetime型
datetime型は、日付と時刻を扱える。

```
apollo_11 = datetime.datetime(1969,7,21,5,17,40)
```

```
print(apollo_11)
```
